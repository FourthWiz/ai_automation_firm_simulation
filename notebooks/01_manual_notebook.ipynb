```python
import numpy as np
from firm_ai_abm.config import FirmParams
from firm_ai_abm.firm import make_firm
from firm_ai_abm.production import Mode
```

```python
param_dict = {
    "seed": 42,
    "N": 1000, # I want to do this variable. for example firm firing people and reduces N if needed
    "T": 60,
    "tasks_per_worker": 10, # Here I want to make this normally distributed - so that there are some low performers
    "q_h": 1.0, # Ok, this will be like normalized tasks
    "q_a": 1.2, # Here assume for perfectly automatable task machine makes it better  
    "g": 0.5, # This is addition from augmentation
    "w": 1.0, # Wage per worker, here in future I want to have it variable and correlated with tasks per worker. But not perfectly
    "c_aug": 0.05, #  Cost of augmentation per task
    "c_auto": 0.4, # automation infrastructure cost per task
    "c_fire": 2.0, # Cost of firing a worker
    "c_hire": 0.5, # cost of hiring a worker
    "c_train": 0.1, # cost of training a worker one time. It's said switching H->A, but not sure if it's a constant mode. It might be one time
    #So as soon as the worker is trained he no longer needs training and can perform A tasks.
    "F": 5.0, # Fixed cost of running the firm, e.g. rent, utilities, etc.
    "p": 1.0, # Price per unit of output. Should be at least with some margin above the cost of production. Currently if w=1.0, and 
    # tasks per worker = 10, then cost per task is 0.1, so price should be above that. I think 1.0 is too large. but for now ok.
    "n_amortize": 6, # Number of periods to amortize automation investment. So if we invest c_auto per task, 
    #then we can consider that cost as c_auto/n_amortize per period for T periods. 
    # This is a simplification, but it allows us to model the investment in automation over time.
    #  In reality, the cost might be amortized differently, but this is a reasonable assumption for our model.    
}
```

```python
firm = make_firm(FirmParams(**param_dict))
```
