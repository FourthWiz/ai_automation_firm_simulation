```python
import numpy as np
from firm_ai_abm.config import FirmParams
from firm_ai_abm.firm import make_firm
from firm_ai_abm.production import Mode

firm = make_firm(FirmParams(seed=0))
print(f"alpha.shape={firm.alpha.shape} dtype={firm.alpha.dtype}")
print(f"beta.shape={firm.beta.shape} dtype={firm.beta.dtype}")
print(f"K={firm.K}; modes all H? {(firm.modes == int(Mode.H)).all()}")

# R-07: reset() must NOT re-sample alpha/beta
alpha_before = firm.alpha.copy()
beta_before = firm.beta.copy()
firm.modes[:5] = int(Mode.T)  # mutate something resettable
firm.history.append({"t": 0, "pi": -1.0})
firm.reset()
assert np.array_equal(firm.alpha, alpha_before), "R-07 violation: alpha changed across reset()"
assert np.array_equal(firm.beta, beta_before),   "R-07 violation: beta changed across reset()"
assert (firm.modes == int(Mode.H)).all(),        "reset() did not restore all-H modes"
assert firm.K == 10,                              "reset() did not restore K=N/tasks_per_worker"
assert firm.history == [],                        "reset() did not clear history"
print("R-07 reset() check: PASS")
```
